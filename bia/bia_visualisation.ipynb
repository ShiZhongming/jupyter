```python
import numpy as np
import pandas as pd
import calmap

# path to the file created in the previous step
path = r'.../outputs/data/potentials/agriculture/plot/B1000_BIA_planting_calendar_AmaranthRed.csv' 

# read the file
df = pd.read_csv(path)

# select the column of values to plot; this example the plots values in the 26th column
df_to_visual = df.iloc[3:,[25]]

# create a series with the dates as index; in this example, the 26th column has the column name: 24
# this column represents Floor number: 1 and Orientation: South
days = pd.date_range('1/1/2022', periods=365, freq='D')
events = pd.Series(df_to_visual['24'].values.astype(float), index=days) 

# plot and format
calmap.calendarplot(events, monthticks=True, daylabels='MTWTFSS', dayticks=True,
                     cmap='PuBuGn', linewidth=2,
                    fig_kws=dict(figsize=(13, 13)))
```

```python
all_days = pd.date_range('1/1/2022', periods=365, freq='D')
days = np.random.choice(all_days, 365)
events = pd.Series(np.random.randn(len(days)), index=days)
```
